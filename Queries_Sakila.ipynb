1. What query would you run to get all the customers inside city_id=312?

```python
q ="""SELECT c.first_name, c.last_name, c.email, a.address, city.city
FROM customer AS c
JOIN address AS a
ON a.address_id = c.address_id
JOIN city
ON a.city_id = city.city_id
WHERE city.city_id =312;"""
pd.read_sql(q,engine)
```

2. What query would run to get all comedy films?

```python
q="""SELECT  f.title, f.description, f.release_year, f.rating,
f.special_features, c.name
FROM film AS f
JOIN film_category
ON film_category.film_id = f.film_id
JOIN category AS c
ON c.category_id = film_category.film_id
WHERE name = 'comedy';"""
pd.read_sql(q,engine)
```

3. What query would you run to get all films that Johnny Lollobrigida was in?

```python
q="""SELECT  a.last_name, f.title, f.release_year
FROM actor AS a
JOIN film_actor
ON film_actor.actor_id = a.actor_id
JOIN film as f
ON f.film_id = film_actor.film_id
WHERE last_name = 'Lollobrigida';"""
pd.read_sql(q,engine)
```

What query would you run to get the first and last names of all the actors in the movie titled "Bingo Talented"?

```python
q="""SELECT a.first_name, a.last_name, f.title
FROM actor AS a
JOIN film_actor
ON film_actor.actor_id = a.actor_id
JOIN film as f
ON f.film_id = film_actor.film_id
WHERE title = 'Bingo Talented';"""
pd.read_sql(q, engine)
```

5. What query would you run to get the customer_id associated with all payments great than twice the average payment amount?

```python
q= """SELECT customer_id, amount
FROM payment
WHERE amount > 2 * (SELECT AVG(amount) FROM payment);"""
pd.read_sql(q,engine)
```

6. What query would you run to list the first and last names of the 5 customers who have the highest number(count) of payments? 

```python
q = """SELECT   c.first_name, c.last_name, p.payment_id,
	(SELECT COUNT(p.payment_id) AS num_payment
	FROM payment as p
	WHERE c.customer_id = p.customer_id)
FROM customer as c
ORDER BY num_payments DESC
LIMIT 5;
"""
pd.read_sql(q, engine)
```
